```python
def read_csvtagged(csvtagged_path):
    """
    Reads rows from a csv .tagged file.
    Each row consists of 3 columns of information:

    COLUMN	DESCRIPTION
    ID	Unique ID for this datapoint
    TEXT	Two snippets of text separated by [SNIPPET]
    LABEL	The label for this datapoint (see below)

    The labels are:
    0	Not the same author
    1	Same author
    """
    rows = []
    with open(csvtagged_path, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            rows.append((row[0], row[1], row[2]))
    return rows


# read in csv training data file
everything = read_csvtagged('train.csv')[1:]


everything_x = [text for _,text,_ in everything]
# everything_y = [int(label) for _,_,label in everything]
ids = [id for id,_,_ in everything]


def split_text(text_list):
    #split lines of text on [SNIPPET] and return lists of first and second excerpts
    texta = []
    textb = []
    for line in text_list:
        a, b = line.split("[SNIPPET]")
        texta.append(a)
        textb.append(b)
    return texta, textb


train_text_a, train_text_b = split_text(everything_x)

snippet_lengths = [len(text) for text in train_text_a+train_text_b]
avg_snippet = mean([len(text) for text in train_text_a+train_text_b])
print(f"Average snippet length: {avg_snippet}")
# min_snippet_a = min([len(text) for text in train_text_a])
print(f"Min snippet length: {min(snippet_lengths)}")
# max_snippet_a = max([len(text) for text in train_text_a])
print(f"Max snippet length: {max(snippet_lengths)}")
```
